Let's make up some data that we already know is normally distributed

```python
import numpy as np
random_normals = np.random.randn(1000)
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create a Q-Q lot
sm.qqplot(random_normals, line='45')
```

```python
sm.qqplot(random_normals, line='45')
```

```python
sm.qqplot(random_normals, line='q')
```

Data falls on the straight line. So, our data came from a normal distribution which we already knew

Let's take a look at the Q-Q plot for something that's not normal

```python
random_uniform = np.random.rand(1000)
sm.qqplot(random_uniform, line='45')
```
